```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
check_lengths = pd.read_csv('C:/Users/kiril/Desktop/test/tinkoff_project_data_analysis_predict_check_lengths_pred_len.csv')
check_lengths['person_id'] = check_lengths['Unnamed: 0']
check_lengths = check_lengths[['pred_len','person_id']]

# check_lengths длина чека 

def listfromstr1(stroka):
    return np.fromstring(stroka[1:len(stroka)-1:1], dtype=int, sep=', ')

menu = pd.read_csv('C:/Users/kiril/Desktop/test/menu1.csv')
menu['good_id'] = menu['good_id'].apply(listfromstr1)

# menu меню

prediction = pd.read_csv('C:/Users/kiril/Desktop/test/prediction.csv')
middle_prediction = pd.read_csv('C:/Users/kiril/Desktop/test/middle_prediction.csv')

# prediction рейтинг пользователей

def get_top_n(df2): # функция делающая предсказания
    df1 = prediction[prediction['person_id'] == df2['person_id']]
    c = menu[(menu['month'] == df2['month']) & (menu['day'] == df2['day'])]['good_id'].values[0]
    arr = check_lengths[check_lengths['person_id'] == df2['person_id']]['pred_len'].values
    if (arr.size > 0):
        n = arr[0]
    else:
        n = 6
        return middle_prediction[np.isin(middle_prediction['good_id'],c)][:n]['good_id'].values
    return df1[np.isin(df1['good_id'],c)][:n]['good_id'].values
```

```python
test = pd.read_csv('C:/Users/kiril/Desktop/test/test.csv')
```

```python
%%time
pred = test.apply(get_top_n, axis=1)
```

```python
end = pd.DataFrame()
end['chknum'] = test['chknum']
end['pred'] = pred
```
