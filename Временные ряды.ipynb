# Временные ряды

```python
import pandas as pd
import numpy as np
from pandas import DataFrame

def listfromstr(stroka):
    return np.fromstring(stroka[1:len(stroka)-1:1], dtype=int, sep=',')
def good_0(good_id):
    return listfromstr(good_id).tolist().count(8)
data = pd.read_csv('C:/Users/kiril/Desktop/td/train_with_menu.csv')

data_need = pd.DataFrame()
data_need['month'] = data['month']
data_need['day'] = data['day']
data_need['good_id'] = data['good_id']

data_need['good_id'] = data['good_id'].apply(good_0)
data_suka = data_need.groupby(['month','day']).good_id.sum()
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,5)
data_suka.groupby(['month','day']).mean().plot(kind='bar')
```
